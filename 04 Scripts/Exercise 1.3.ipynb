```python
import pandas as pd
import numpy as pd
```

```python
# Example DataFrame
df = pd.DataFrame({
    "Year": [1901, 1910, 1920, 1930, 1940],    # numeric column
    "Event": ["Event A", "Event B", "Event C", "Event D", "Event E"]  # text column
})

# Display the DataFrame
df
```

```python
df["Year_plus_10"] = df["Year"].apply(lambda x: x + 10)
df["Year_times_2"] = df["Year"].apply(lambda x: x * 2)
df["Year_minus_5"] = df["Year"].apply(lambda x: x - 5)
df
```

```python
df["Flag"] = df["Event"].apply(lambda x: "Yes" if x in ["Event A", "Event B"] else "No")
df
```

```python
df["Flag_Boolean"] = df["Year"].apply(lambda x: True if x >= 1920 else False)
df

```
