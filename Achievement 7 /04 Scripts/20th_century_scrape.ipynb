```python
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)  # Warnings

url = "https://en.wikipedia.org/wiki/Key_events_of_the_20th_century"

# Use verify=False to avoid SSL certificate errors (acceptable for coursework)
response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"}, verify=False)
response.raise_for_status()

# Parse HTML
soup = BeautifulSoup(response.text, "html.parser")

# Narrow down to the main article content
content = soup.find("div", id="mw-content-text")

# Extract text paragraphs
paragraph_elements = content.find_all("p")
paragraph_texts = [p.get_text(strip=True) for p in paragraph_elements if p.get_text(strip=True)]

# Extract section headings (h2, h3)
heading_elements = content.find_all(["h2", "h3"])
heading_texts = [h.get_text(strip=True) for h in heading_elements]

# Combine text for later processing
all_text = heading_texts + paragraph_texts

# Show some of what we got
for i, text in enumerate(all_text[:15], 1):
    print(f"{i}. {text[:120]}…")
```

```python
# Save scraped content to a text file
file_name = "key_events_20th_century.txt"

with open(file_name, "w", encoding="utf-8") as f:
    for line in all_text:
        f.write(line + "\n\n")

print(f"Scraped content saved to {file_name}")
```
